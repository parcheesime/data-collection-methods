This tutorial will scrape the top 100 ebooks for the day at gutenburg.com and schedule a daily run of the script.

Add libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import urllib3
import re
```

Get data fron link

```python
url = 'https://www.gutenberg.org/browse/scores/top'
html_scraped = requests.get(url)
soup = BeautifulSoup(html_scraped.text, 'html.parser')
```

```python
# View html links
soup.find_all(re.compile('^a'))
```

Search html

```python
file_type='ebooks'

for link in soup.find_all(re.compile('^a'), limit=15):
    print(link)   
```

Clean up text, save as list of lists

```python
file_type='ebooks'
books=[]
for link in soup.find_all(re.compile('^a')):
    if file_type in link.get('href'):
        books.append(link.text.strip())  
```

Create a dataframe of top 100 ebooks

```python
df=pd.DataFrame(books, columns=['top_ebooks'])
```
