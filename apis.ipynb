## Web scraping using APIs

If the site has an API, use it.

```python
# Add libraries
import requests
import json
import urllib3
import pprint as pp
import pandas as pd
```

```python
# function to grab data from a resource url
def getdata(url):
    try:
        response=requests.get(url)
        return response.json()
    except:
        return print(f'A problem occured with your link, status code {response.status_code}')
```

```python
# get data from ca.gov api explorer
getdata('https://caltrans-gis.dot.ca.gov/arcgis/rest/services/CHhighway/CCTV/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json')
```
